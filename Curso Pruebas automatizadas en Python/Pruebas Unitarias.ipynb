# Pruebas Unitarias

Son pruebas automatizadas las cuales tienen como objetivo verificar el funcionamiento de una unidad de codigo. Una unidad puede ser la parte mas pequena de un codigo de programacion.

Cada funcion puede definirse como unidades

Para las pruebas es necesario utilizar un metodo de confirmacion ya sea utilizando Doctest o cualquier terminal que nos permita realizar las operaciones. Claro esta que en este caso utilizamos jupyternotebook en el cual podemos testear a nivel basico


## Doctest

```python
import doctest
```

```python
def fibonacci(number):
    if number == 0:return 0
    elif number == 1: return 1
    else:
        return fibonacci(number - 1)+ fibonacci(number-2)
'''
>>> fibonacci(12)
144

>>> fibonacci(16)
987
'''
doctest.testmod()
```

```python
def palindromo(sentence):
    sentence = str(sentence.lower().replace(" ", ""))
    return sentence == sentence[::-1]
'''Retorna verdadero si el parametro es un palidromo
en caso contrario retornara falso

sentence == String o entero
>>> palindromo("anita lava la tina")
True
>>> palindromo('12321')
True
'''
doctest.testmod()
```

```python



class Recursivo:
    def factorial(self, number):
        if number == 0: return 1
        else: return number * self.factorial(number-1)

'''
>>> recursivo = Recursivo()
>>> recursivo.factorial(5)
120

>>> recursivo.factorial(13)
6227020800

'''
doctest.testmod()
```

Listado de métodos asserts que podemos utilizar con la librería unittest.

    Si tú no te encuentras usando dicha librería, puedes utilizar la palabra reservada assert en conjunto con la columna Evalua para poder realizar tus pruebas.
    
   
Método	Evalua	Versión

assertEqual(a, b)	a == b	> 2.7

assertNotEqual(a, b)	a != b	> 2.7

assertTrue(x)	bool(x) is True	> 2.7

assertFalse(x)	bool(x) is False	> 2.7

assertAlmostEqual(a, b)	round(a-b, 7) == 0	> 2.7

assertNotAlmostEqual(a, b)	round(a-b, 7) != 0	> 2.7

assertIs(a, b)	a is b	> 3.1

assertIsNot(a, b)	a is not b	> 3.1

assertIsNone(x)	x is None	> 3.1

assertIsNotNone(x)	x is not None	> 3.1

assertIn(a, b)	a in b	> 3.1

assertNotIn(a, b)	a not in b	> 3.1

assertIsInstance(a, b)	isinstance(a, b)	> 3.1

assertNotIsInstance(a, b)	assertNotIsInstance(a, b)	> 3.1

assertGreater(a, b)	a > b	> 3.1

assertGreaterEqual(a, b)	a >= b	> 3.1

assertLess(a, b)	a < b	> 3.1

assertLessEqual(a, b)	a <= b	> 3.1

assertRegex(s, r)	r.search(s)	> 3.1

assertNotRegex(s, r)	not r.search(s)	> 3.2

assertNotRegex(s, r)	len(s) == len(r)	> 3.2

# Covertura
con esto podemos testear si nuestro codigo esta siendo utilizado, debe ser instalado. seguido de esto se debe colocar

    coverage report script.py
    
    coverage run script_test.py
    
Esto creara un archivo y debe ser ejecutado nuevamente el report para visualizar el cover que tiene este mismo. Uilizando el -m para poder visualizar donde no se cubre.

    coverage report script.py -m
    
 Con esto nos podremos decidir que hacer si eliminar o recodificar para probar el codigo con las pruebas unitarias.
 
 
 
 el coverage puede exportarse de diversas maneras utilizando luego del run del script_test.py
 
     coverage html script.py
 
 Si se quiere encender un servidor http para poder leer los archivos se utiliza el siguiente comando.
 
     python3 -m http.server
